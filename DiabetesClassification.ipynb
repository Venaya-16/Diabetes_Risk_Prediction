```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
Categorical=["Sex","Ethnicity","Physical_Activity_Level","Alcohol_Consumption","Smoking_Status"]
for col in Categorical:
  df[col]=encoder.fit_transform(df[col])
```

```python
df.head()
```

```python
Y=df["Family_History_of_Diabetes"]
X=df.drop("Family_History_of_Diabetes",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model=DecisionTreeClassifier(random_state=42)
model.fit(X_train,Y_train)
```

```python
Y_predict=model.predict(X_test)
accuracy=accuracy_score(Y_test,Y_predict)
print("Accuracy:",accuracy)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(Y_test,Y_predict)
print(cm)
sns.heatmap(cm,annot=True,cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print(classification_report(Y_test,Y_predict))
```

```python
important= pd.DataFrame({"Feature":X.columns,"Importance":model.feature_importances_})
importance=important.sort_values(by="Importance",ascending=False)
```
